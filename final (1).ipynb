```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={'student_id':[1,2,3,4,5],"name":['Alice','Bob','Charlie','Diana','Eve'],
     'math_score':[85,65,95,50,75],'science_score':[90,70,92,55,80],
      'english_score':[88,60,90,45,78]}
df=pd.DataFrame(data)
print(df)
```

```python
df = df.dropna()
```

```python
df['average_score'] = df[['math_score', 'science_score', 'english_score']].mean(axis=1)
print(df)
```

```python
subject_averages = df[['math_score', 'science_score', 'english_score']].mean(axis=0)
print("Subject Averages:", subject_averages)
```

```python
top_students = df.sort_values(by='average_score', ascending=False)
print("Top 5 Students:", top_students[['name', 'average_score']])
```

```python
top_students.plot(x='name', y='average_score', kind='bar', title='Top 5 Students by Average Score')
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.show()
```

```python
subject_averages.plot(kind='bar', title='Average Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.show()
```
